```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
# create DataFrame
df = pd.DataFrame({'NSS feature extraction': [ 2.77, 4.91, 8.66],
                   'Image projection': [ 0.46198153495788574, 0.6406234097480774, 0.936877965927124],
                   'Inference': [0.7602, 0.7684, 1.578]},
                  index=['Min', 'Mean', 'Max'])



palette = sns.color_palette("Purples", len(df.columns))

df.plot(kind='bar', stacked=True, color=palette) 

# labels for x & y axis
plt.ylabel('Total time(s)')
 
# title of plot
plt.title('Expected total time for a random point cloud')
plt.xticks(rotation=0)
```
